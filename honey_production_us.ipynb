<p> Honey Production in the USA (1998-2012)</p>

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
#read csv by pandas
df = pd.read_csv('honeyproduction.csv')
```

```python
#inspect data
print(df.head())
```

```python
# I'm only going to focus on the totalproduction and year columns
#make a new df which is average production per year
prod_per_year = df.groupby(['year']).totalprod.mean().reset_index()
```

```python
#Define Xvalue and y value for making the scatter plot
X = prod_per_year.year.values.reshape(-1,1)
y = prod_per_year.totalprod.values
plt.scatter(X,y)
plt.show()
plt.close()
```

```python
# create linear regression model 
#find the model that is going to fit 
regr = linear_model.LinearRegression()
regr.fit(X,y)
print(regr.coef_[0])
y_predict = regr.predict(X)
```

```python
plt.scatter(X,y)
plt.plot(X,y_predict)
plt.xlabel('Year')
plt.ylabel('Average production')
plt.title('Average production of Honey per year')
plt.show()
plt.close()
```

```python
#Let’s predict what the year 2050 may look like in terms of honey production.
X_future = np.array(range(2013,2050))
X_future =X_future.reshape(-1,1)
future_predict = regr.predict(X_future)
plt.scatter(X,y)
plt.plot(X,y_predict)
plt.plot(X_future,future_predict)
plt.xlabel('Year')
plt.ylabel('Average production')
plt.title('Average production of Honey per year')
plt.show()
plt.close()
```

As you can see, honey production will be down to 300,000 lbs by about 2050.
This is a 94% decrease compared to the 2000 figure.
